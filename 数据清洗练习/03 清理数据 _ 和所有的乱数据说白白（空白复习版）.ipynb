## 对列进行拆分

```python
# 读取示例数据
df2 = pd.read_csv("example5.csv")
df2
```

```python
# 任务二：
# 把df2的人口密度列进行拆分，使得最终df2包含“城市”、“人口”、“面积”三列
df2[["人口","面积"]] = df2["人口密度"].str.split("/",expand = True)
df2 = df2.drop("人口密度",axis = 1)
df2
```

## 把不同列合并成一列

```python
df3 = pd.DataFrame({'姓': ['张', '李', '王'], '名': ['三', '四', '五'], '年龄': [30, 25, 35]})
df3
```

```python
# 任务三：
# 把df3的姓和名两列进行合并，得到“姓名”列，并且使得最终df3包含“年龄”、“姓名”两列
df3["姓名"] = df3["姓"].str.cat(df3["名"],sep = "-")
df3 = df3.drop(["姓","名"],axis = 1)
df3
```

## 把宽数据转换成长数据

```python
# 读取示例数据
df4 = pd.read_csv("example6.csv")
df4
```
